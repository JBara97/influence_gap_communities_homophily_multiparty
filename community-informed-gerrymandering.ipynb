```python
size=10000
degs = np.arange(20)
counts_tot = {n:0 for n in range(20)}
counts_tot_er = {n:0 for n in range(20)}
counts_tot_ba = {n:0 for n in range(20)}
for i in trange(size):
    g = nx.relaxed_caveman_graph(2,10,0.1)
    g1 = nx.gnp_random_graph(20,9/20)
    g2 = nx.barabasi_albert_graph(20,7)
    deg,counts=np.unique([d for n,d in g.degree()],return_counts=True)
    deg1,counts1=np.unique([d for n,d in g1.degree()],return_counts=True)
    deg2,counts2=np.unique([d for n,d in g2.degree()],return_counts=True)
    for i,d in enumerate(deg):
        counts_tot[d]+=counts[i]
    for i,d in enumerate(deg1):
        counts_tot_er[d]+=counts1[i]
    for i,d in enumerate(deg2):
        counts_tot_ba[d]+=counts2[i]
```

# Weak Party Assignment (WPA), Fixed Graph (FG)

```python
size=1000
win = np.zeros(size)
lose = np.zeros_like(win)
g_cave2 = nx.relaxed_caveman_graph(2,10,0.1,seed=11)

for i in trange(size):
    af.random_assignment(g_cave2,2)
    vals = list(af.party_assortment(g_cave2,half_to_zero=False).values())
    win[i] = max(vals)
    lose[i] = min(vals)
```

```python
print('Winner: ',np.mean(win))
print('Loser: ', np.mean(lose))
print('Influence gap: ', np.mean(win) - np.mean(lose))
```

```python
fig_weak=plt.figure(figsize=(10,8))
ax=plt.gca()
ax.scatter(range(size),win,label='Winner',c='red',marker='+',alpha=0.5)
ax.scatter(range(size),lose,label='Loser',c='blue',marker='x',alpha=0.5)
# ax.scatter(range(size),win-lose,label='Gap',c='k',marker='o',alpha=0.5)
ax.plot([0,size-1],[np.mean(win)]*2,c='k',lw=4.5)
ax.plot([0,size-1],[np.mean(win)]*2,c='red',lw=3,label=r'$\mathcal{A}_W = $%.3f'%np.mean(win))
ax.plot([0,size-1],[np.mean(lose)]*2,c='k',lw=4.5)
ax.plot([0,size-1],[np.mean(lose)]*2,c='blue',lw=3,label=r'$\mathcal{A}_L = $%.3f'%np.mean(lose))
ax.legend(loc='lower right',framealpha=0.9)
ax.set_xlabel("Random party assignment")
ax.set_ylabel("Influence Assortment")
ax.set_title("Party Influence for weak party assignment, fixed graph")
txt = "For a fixed relaxed-caveman-graph, we allocate parties using WPA,"
txt += 'resulting in an influence gap of $\mathcal{G}_I = %.3f$'%(np.mean(win)-np.mean(lose))
# fig_weak.text(.5, 0, txt, ha='center')
# fig_weak.tight_layout()
```

```python
fig_weak.savefig("Party-influence-WPA-FG.pdf",bbox_inches='tight')
```

```python
size=10000
IG = np.zeros((size,4)) # Columns = ['WPA-FG','WPA-RG','SPA-FG','SPA-RG'])
PIA = np.zeros((size,2,4))
node_list_spa = np.random.choice(range(20),size=10,replace=False) #SPA-RG

g_sf = nx.relaxed_caveman_graph(2,10,0.1,seed=11) #SPA-FG

g_wf = g_sf.copy() #WPA-FG

# Note that nodes {0,..,9} are in community 1, 
    # and {10,...,19} are in community 2 always

for i in trange(size):
    g_sr = nx.relaxed_caveman_graph(2,10,0.1) #SPA-RG
    g_wr = g_sr.copy() #WPA-RG
    node_list_sparg = np.random.choice(range(20),size=10,replace=False) #SPA_FG
    parties_fg = np.random.randint(2,size=20) #WPA-FG
    
    af.dual_party_assignment(g_sr,node_list_spa) #SPA-RG
    af.dual_party_assignment(g_sf,node_list_sparg) #SPA-FG
    
    nx.set_node_attributes(g_wr,parties_dict,'party') #WPA-RG
    af.random_assignment(g_wf,2) #WPA-FG
    
    IG[i,:] = np.array([np.abs(np.diff(list(af.party_assortment(g_wf).values()))[0]),
               np.abs(np.diff(list(af.party_assortment(g_wr).values()))[0]),
               np.abs(np.diff(list(af.party_assortment(g_sf).values()))[0]),
               np.abs(np.diff(list(af.party_assortment(g_sr).values()))[0])])
    
    PIA[i,:,:] = np.transpose(np.array([np.sort(list(af.party_assortment(g_wf).values())),
                           np.sort(list(af.party_assortment(g_wr).values())),
                           np.sort(list(af.party_assortment(g_sf).values())),
                           np.sort(list(af.party_assortment(g_sr).values()))]))
```

# Systematising hRCs

```python
def hRC_wrapper(size=1000,pas=100,h_size=3,N=20,p0=0.1):
#     size = 100
#     pas = 100
#     h_size = 3
#     N = 20

    IG = np.zeros((size,pas,h_size))
    h = np.linspace(0.5,1,h_size)
    node_list = np.transpose(np.array([np.random.choice(range(N),size=(N//2),replace=False) for i in range(pas)]))
    parties = [{n:int(n in node_list[:,j]) for n in range(N)} for j in range(pas)]
    if (len(np.unique([af.PA_to_dec(node_list[:,j],N) for j in range(pas)]))) == pas:
        g_hRC = [[[af.homophilic_relaxed_caveman_graph(2,N//2,p0,h[k],parties[j]) 
                   for j in range(pas)] for k in range(h_size)] for i in trange(size)]
        for i in range(size):
            for j in range(pas):
                for k in range(h_size):
                    IG[i,j,k] = np.abs(np.diff(list(af.party_assortment(g_hRC[i][k][j]).values()))[0])
                    
    names = ['h', 'PA', 'graph']
    index = pd.MultiIndex.from_product([h,[af.PA_to_hex(node_list[:,j],N) for j in range(pas)],range(size)], names=names)
    df = pd.DataFrame({'IG': IG.flatten()}, index=index)['IG'].unstack(level='PA').sort_index()
    return df

def hex_to_PA(hx,N,party_dict=True):
    bitstr = bin(int(hx,16))[2:].zfill(N)
    out = {n:int(bitstr[n]) for n in range(20)}
    if party_dict:
        return out
    else:
        return np.array(list(out.keys()))[np.array(list(out.values()))==1]
```

# Dynamics

```python
import scipy.io
mat = sp.io.loadmat('dynamic_parameters.mat')
par_dist = np.zeros((3,2,11))
par_dist[0,0,:] = mat['Win1'].reshape(12)[:-1]
par_dist[0,1,:] = mat['Win2'].reshape(12)[:-1]

par_dist[1,0,:] = mat['Deadlock1'].reshape(12)[:-1]
par_dist[1,1,:] = mat['Deadlock2'].reshape(12)[:-1]

par_dist[2,0,:] = mat['Lose1'].reshape(12)[:-1]
par_dist[2,1,:] = mat['Lose2'].reshape(12)[:-1]

def poll_fraction(g,n):
    party = g.nodes[n]['party']
    delta = 1 # Since the node knows it will vote for its own party
    neighbours = list(g.neighbors(n))
    for m in neighbours:
        if g.nodes[m]['party'] == party:
            delta+=1
    delta = delta/(len(neighbours)+1)
    return delta

def dynamics(g,last=False,ass=False,late=False):
    V = 0.6 # Super-majority
    
    T = 240
    t_decision = 3.3
    t = 0
    
    t_transition = 83
    no_decisions = int(T//t_decision)
    
    polls = np.array([poll_fraction(g,n) for n in g])
    initial_parties = np.array(list(nx.get_node_attributes(g,'party').values()))
    intentions = np.zeros((no_decisions+1,len(g)))
    intentions[0,:] = initial_parties
    
    if ass:
        assortment = np.zeros((no_decisions+1,2))
        assortment[0,:] = list(af.party_assortment(g).values())
    
    rand_numbers = np.random.rand(no_decisions+1,len(g))
    
    pars = np.zeros((3,2,len(g)))
    for i in range(3):
        for j in range(2):
            pars[i,j,:] = np.random.choice(np.linspace(0,1,11),p=par_dist[i,j,:],size=len(g))
    
    phase = 0 # Early
    for t_index in range(1,no_decisions+1):
        t += t_decision
#         t_index = int(t//t_decision)
        if t > 83:
            phase = 1 # Late
            if late & (t<83+3.3):
                IG = af.influence_gap(g,absolute=False)
        pars_now = np.zeros(len(g))
        pars_now[polls>=V] = pars[0,phase,polls>=V]
        pars_now[(polls>=1-V)&(polls<V)] = pars[1,phase,(polls>=1-V)&(polls<V)]
        pars_now[polls<1-V] = pars[2,phase,polls<1-V]
        
        for n in g:
            if rand_numbers[t_index,n] <= pars_now[n]:
                intentions[t_index,n] = int(initial_parties[n])
            else:
                intentions[t_index,n] = int(1 - initial_parties[n])
        parties_dict = {n:int(intentions[t_index,n]) for n in g}
        nx.set_node_attributes(g,parties_dict,'party')
        if ass:
            assortment[t_index,:]=list(af.party_assortment(g).values())
#         print(t_index)
#         print(initial_parties)
    if last:
        return intentions[-1,:]
    elif late:
        return IG,np.sum(intentions[-1,:])/len(g)-0.5
    elif ass:
        return assortment
    else:
        return intentions
```

# r surface

```python
N = 20
# ps = np.arange(0.1,1.1,0.1)
# voter_skew = np.zeros((10**4,len(ps)))
# IG = np.zeros((10**4,len(ps)))

p = np.linspace(0.1,1,10)
h = np.linspace(0,1,41)
r = np.zeros((41,10))

for j in trange(41):
    for k in range(10):
        VS = np.zeros(2*10**3)
        IG = np.zeros(2*10**3)
        for i in range(2*10**3):
            node_list = np.random.choice(range(N),size=(N//2),replace=False)
            parties = {n:int(n in node_list) for n in range(N)}    
            G = af.homophilic_relaxed_caveman_graph(2,10,p[k],h[j],parties)
            IG[i] = af.influence_gap(G,absolute=False)
            VS[i] = np.sum(dynamics(G,last=True))/N - 0.5
        r[j,k] = sp.stats.linregress(IG,VS).rvalue
```

```python
df = pd.DataFrame(data=r)
df.to_csv("Dynamics_rcoeff.csv",index=False)
```

```python
P,H = np.meshgrid(p,h)
fig = plt.figure(figsize=(10,10))
ax = plt.axes(projection='3d')
ax.plot_surface(H,P,r,cmap=plt.cm.plasma, linewidth=0.2)

ax.set_xlabel("Homophily",labelpad=7)
ax.set_ylabel("Rewire Probability",labelpad=9)
ax.set_zlabel("Pearson's Correlation",labelpad=15)

fig.savefig("Dynamics_rcoeff.png",bbox_inches='tight')
```
